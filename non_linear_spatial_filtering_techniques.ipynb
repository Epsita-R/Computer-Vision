```python
import cv2
from google.colab.patches import cv2_imshow  # Import cv2_imshow for Colab

# Function to apply median filtering
def median_filter(image, kernel_size):
    return cv2.medianBlur(image, kernel_size)

# Function to apply bilateral filtering
def bilateral_filter(image, d, sigma_color, sigma_space):
    return cv2.bilateralFilter(image, d, sigma_color, sigma_space)

# Function to apply non-local means denoising
def nlmeans_denoising(image, h=10, search_window=21, template_window=7):
    return cv2.fastNlMeansDenoising(image, None, h, search_window, template_window)

# Function to apply adaptive thresholding
def adaptive_thresholding(image, max_value=255, adaptive_method=cv2.ADAPTIVE_THRESH_GAUSSIAN_C, threshold_type=cv2.THRESH_BINARY, block_size=11, c=2):
    return cv2.adaptiveThreshold(image, max_value, adaptive_method, threshold_type, block_size, c)

# Function to apply morphological operation (erosion or dilation)
def morphological_operation(image, operation=cv2.MORPH_CLOSE, kernel_size=(5,5)):
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernel_size)
    return cv2.morphologyEx(image, operation, kernel)

# Read input image
img_path = '/content/Litt_Up_Mug_(4x06).png'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Read grayscale image

# Check if image is loaded successfully
if img is None:
    print(f"Error: Unable to load image '{img_path}'")
    exit()

# Apply Median Filtering
median_filtered_img = median_filter(img, 5)

# Apply Bilateral Filtering
bilateral_filtered_img = bilateral_filter(img, 9, 75, 75)

# Apply Non-local Means Denoising
nlmeans_filtered_img = nlmeans_denoising(img)

# Apply Adaptive Thresholding
thresholded_img = adaptive_thresholding(img)

# Apply Morphological Erosion
eroded_img = morphological_operation(img, cv2.MORPH_ERODE)

# Apply Morphological Dilation
dilated_img = morphological_operation(img, cv2.MORPH_DILATE)

# Display results using cv2_imshow
cv2_imshow(img)
cv2_imshow(median_filtered_img)
cv2_imshow(bilateral_filtered_img)
cv2_imshow(nlmeans_filtered_img)
cv2_imshow(thresholded_img)
cv2_imshow(eroded_img)
cv2_imshow(dilated_img)
```
