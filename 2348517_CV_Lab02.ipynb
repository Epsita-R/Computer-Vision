```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
image = cv2.imread('/content/WhatsApp Image 2023-06-05 at 08.54.15.jpg')
rows, cols = image.shape[:2]
```

#Transformations

Translation

```python
M_translation = np.float32([[1, 0, 70],
                            [0, 1, -40]])
image_translation = cv2.warpAffine(image, M_translation, (cols, rows))

plt.figure(figsize=(12, 12))
plt.subplot(231), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(232), plt.imshow(cv2.cvtColor(image_translation, cv2.COLOR_BGR2RGB)), plt.title('Translation')
plt.tight_layout()
plt.show()
```

Rotation

```python
angle = 50
M_rotation = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
image_rotation = cv2.warpAffine(image, M_rotation, (cols, rows))

plt.figure(figsize=(12, 12))
plt.subplot(231), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(232), plt.imshow(cv2.cvtColor(image_rotation, cv2.COLOR_BGR2RGB)), plt.title('Rotation')
plt.tight_layout()
plt.show()
```

Scaling

```python
M_scaling = np.float32([[1.25, 0, 0],
                        [0, 1.75, 0]])
image_scaling = cv2.warpAffine(image, M_scaling, (cols, rows))

plt.figure(figsize=(12, 12))
plt.subplot(231), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(232), plt.imshow(cv2.cvtColor(image_scaling, cv2.COLOR_BGR2RGB)), plt.title('Scaling')
plt.tight_layout()
plt.show()
```

Shearing

```python
M_shearing = np.float32([[1, 0.5, 0],
                         [0.5, 1, 0]])
image_shearing = cv2.warpAffine(image, M_shearing, (cols, rows))

plt.figure(figsize=(12, 12))
plt.subplot(231), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(232), plt.imshow(cv2.cvtColor(image_shearing, cv2.COLOR_BGR2RGB)), plt.title('Shearing')
plt.tight_layout()
plt.show()
```

Reflection (Horizontal)

```python
M_reflection_horizontal = np.float32([[-1, 0, cols],
                                      [0, 1, 0]])
image_reflection_horizontal = cv2.warpAffine(image, M_reflection_horizontal, (cols, rows))

plt.figure(figsize=(12, 12))
plt.subplot(231), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(232), plt.imshow(cv2.cvtColor(image_reflection_horizontal, cv2.COLOR_BGR2RGB)), plt.title('Reflection (Horizontal)')
plt.tight_layout()
plt.show()
```

Reflection (Vertical)

```python
M_reflection_vertical = np.float32([[1, 0, 0],
                                    [0, -1, rows]])
image_reflection_vertical = cv2.warpAffine(image, M_reflection_vertical, (cols, rows))

plt.figure(figsize=(12, 12))
plt.subplot(231), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(232), plt.imshow(cv2.cvtColor(image_reflection_vertical, cv2.COLOR_BGR2RGB)), plt.title('Reflection (Vertical)')
plt.tight_layout()
plt.show()
```
